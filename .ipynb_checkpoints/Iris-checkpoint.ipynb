<h1>Iris Dataset - Supervised and Unsupervised Learning</h1>

<h2>Part 1 : Data Extraction</h2>

```python
# import necessary modules
from sklearn.datasets import load_iris
import pandas as pd
from pandas.plotting import scatter_matrix
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.cluster import KMeans
```

```python
# set style of plots as ggplot
plt.style.use('ggplot')
```

```python
# load data
iris = load_iris()
```

```python
# check type of data
type(iris)
```

```python
# print keys of Iris dataset
print(iris.keys())
```

```python
type(iris.data), type(iris.target)
```

```python
# check dimensions of data
iris.data.shape
```

```python
# check dimensions of target
iris.target.shape
```

```python
# display target classes
iris.target_names
```

<h2>Part 2 : Exploratory Data Analysis (EDA)</h2>

```python
# reassign data and target tuples to new values X & y respectively
X = iris.data
y = iris.target
```

```python
# Using Pandas, construct a DataFrame that 
# consists of columns containing the values of each feature 
# and is labelled by the name of that feature
df = pd.DataFrame(X, columns = iris.feature_names)
```

```python
# View the first five rows of the data
df.head()
```

```python
# Construct a scatter matrix to check correlation between various columns
_ = scatter_matrix(df, c = y, figsize = [8, 8], s = 150, marker = 'D')
```

<h3>Note : We can observe a decent correlation between petal length and petal width</h3>

<h2>Part 3 : Supervised Learning (Classification using K-Nearest Neighbours Classifier)</h2>

```python
# create an instance of the KNN classifier
knn = KNeighborsClassifier(n_neighbors = 6)
```

```python
# split the given data into training and validation sets
# 'random_state' ensures random splitting of tuples
# 'stratify' ensures that balance in the count of target values
# is maintained in both training and validation datasets after split 
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.3, random_state = 21, stratify = y)
```

```python
# fit the model to the data to train the classifier
knn.fit(X_train, y_train)
```

```python
# use trained classifier to predict values on validation data
y_pred = knn.predict(X_val)
```

```python
print("Validation Set Predictions : {}".format(y_pred))
```

```python
# check accuracy of KNN model
knn.score(X_val, y_val)
```

```python
# display confusion matrix for the given classifier
confusion_matrix(y_val, y_pred)
```

```python
# produce classification report for KNN classifier
print(classification_report(y_val, y_pred))
```

<h2>Part 4 : Unsupervised Learning (Clustering using K-Means Clustering Algorithm)</h2>

```python
# instantiate a kmeans clustering model
model = KMeans(n_clusters = 3)
```

```python
# fit the model to the data to being training
model.fit(X)
```

```python
# predict labels for sample data using trained model
labels = model.predict(X)
```

<h3>Visualize clustering of sample points using a Scatter Plot where :<br><br>
X axis : Petal Length <br><br>
Y axis : Petal Width</h3>

```python
# assign petal length attribute to variable 'xs'
xs = X[:, 0]
```

```python
# assign petal width attribute to variable 'ys'
ys = X[:, 2]
```

```python
# generate scatter plot
plt.scatter(xs, ys, c = labels)
```
